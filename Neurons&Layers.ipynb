Logistic Neuron : We can implement a 'logistic neuron' by adding a sigmoid activation. Tensorflow is most often used to create multi-layer models. The Sequential model is a convenient means of constructing these models.

```python
model = Sequential(
    [
        tf.keras.layers.Dense(1,  activation = 'sigmoid',input_dim=1,  name='L1')
    ]
)
```

```python
model.summary()
```

```python
logistic_layer = model.get_layer('L1')
w,b = logistic_layer.get_weights()
print(w,b)
print(w.shape,b.shape)
```

```python
set_w = np.array([[2]])
set_b = np.array([-4.5])
# set_weights takes a list of numpy arrays
logistic_layer.set_weights([set_w, set_b])
print(logistic_layer.get_weights())
```

```python
a1 = model.predict(X_train[0].reshape(1,1))
print(a1)
```
